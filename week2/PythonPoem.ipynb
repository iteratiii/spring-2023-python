```python
# 10)

# Original poem from https://www.poetryfoundation.org/poems/47536/one-art

title = '''One Art
by Elizabeth Bishop
elided by Gabriel Lee
'''

losing = '''The art of losing isn't hard to master;
so many things seem filled with the intent
to be lost that their loss is no disaster.

Lose something every day. Accept the fluster
of lost door keys, the hour badly spent.
The art of losing isn't hard to master.

Then practice losing farther, losing faster:
places, and names, and where it was you meant
to travel. None of these will bring disaster.

I lost my mother's watch. And look! my last, or
next-to-last, of three loved houses went.
The art of losing isn't hard to master.

I lost two cities, lovely ones. And, vaster,
some realms I owned, two rivers, a continent.
I miss them, but it wasn't a disaster.

—Even losing you (the joking voice, a gesture
I love) I shan't have lied. It's evident
the art of losing's not too hard to master
though it may look like (Write it!) like disaster.

'''
# ’

print(title)
print(losing)

# 20k.txt is a list of the 10,000 most common English words in order of frequency, 
# as determined by n-gram frequency analysis of the Google's Trillion Word Corpus.
# From https://github.com/first20hours/google-10000-english

# Open the list of words as a file, and line by line, make it into a list, 
# using rstrip() method to remove any trailing characters.

with open("20k.txt") as file:
    most_common_words = [line.rstrip() for line in file]
    
# Turn the list into only the first n items. 
# Is there a more elegant way to do this, just by importing the first n?
most_common_words = most_common_words[:1500]

# Check that this worked properly and the list is n words long
# print(len(most_common_words))
# Flip da list

most_common_words.append("I")
most_common_words.reverse()

# remove all one-character words
for word in most_common_words:
    if(len(word) == 1 and word != "a" and word != "i"):
        most_common_words.remove(word)

# Check how many words remain (976)
# print(len(most_common_words))

# For every word in list, create a spacer that's only spaces, the same length as the word
# Create a dict 
# Dict keys = strings to test — if it's not part of another word, it should be b/t spaces, newline, or punctuation
    # There is probably a smarter regex way to do this
# Dict values = strings to replace in the same order, using the spacer
# For every string in the test list, look at the lowercased poem
# If found, take the test key and replace with spacer value
# https://www.geeksforgeeks.org/python-test-if-string-contains-element-from-list/

lost_words = []

import string 
# print(string.punctuation)

for word in most_common_words:
    spaces = " "*len(word)
    # Creating the test_replace dictionary
    test_replace = {" " + word + " ":" " + spaces + " ",
                    "\n" + word + " ":"\n" + spaces + " ",
                    "\n" + word + " ":"\n" + spaces + " ",
                    " " + word + "\n":" " + spaces + "\n"}
    for punct in string.punctuation:
        test_replace.update({punct + word + " ":punct + spaces + " ",
                        " " + word + punct:" " + spaces + punct,
                        punct + word + "\n":punct + spaces + "\n",
                         punct + word + "\n":punct + spaces + "\n",
                        "\n" + word + punct:"\n" + spaces + punct,
                        "\n" + word + punct:"\n" + spaces + punct})
    for test in test_replace:
        if(losing.lower().find(test) != -1):
            losing = losing.replace(test, test_replace[test])
            lost_words.append(word)
            print(losing)

            
# Convert lost_words list to dict with keys to list again, instead of a set, to remove dupes but preserve order
# https://stackoverflow.com/questions/480214/how-do-i-remove-duplicates-from-a-list-while-preserving-order
# Print it out
lost_words_string = ', '.join(list(dict.fromkeys(lost_words)))
print ("\n(Words lost, in order of loss: " + lost_words_string + ")")
```
